```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('laufzeitmessungen.csv', sep=";", index_col=0)
```

```python
data.head()
```

## Get a first impression of the runtime behavior by plotting the data.

| x axis 	| y axis 	| growth   	|
|--------	|--------	|----------	|
| linear 	| linear 	| Linear   	|
| log    	| linear 	| O(log n) 	|
| linear 	| log    	| O(2^n)   	|
| log    	| log    	| O(n^a)   	|

## Algorithm 1

#### Both axes are linear

```python
time = range(1, len(data['T1(n)'])+1)
plt.plot(data['T1(n)'], time)
plt.xlabel('Algorithm 1')
plt.show()
```

#### Log of both axes

```python
plt.plot(np.log(data['T1(n)']), np.log(time))
plt.xlabel('Algorithm 1')
plt.show()
```

#### Log of x axis only

```python
plt.plot(np.log(data['T1(n)']), np.log(time))
plt.xlabel('Algorithm 1')
plt.show()
```

#### Log of y axis only

```python
plt.plot(data['T1(n)'], np.log(time))
plt.xlabel('Algorithm 1')
plt.show()
```

*So the algorithm should have a exponential complexity O(2^n)*

## Algorithm 2

```python
time = range(1, len(data['T2(n)'])+1)
plt.plot(data['T2(n)'], time)
plt.xlabel('Algorithm 2')
plt.show()
```
